```python
# for grouping the base flavour is perfect.
group_by_frame = df.groupby('Base Flavor')
```

```python
group_by_frame[['Flavor Rating','Texture Rating','Total Rating']].mean()
```

```python
numeric_columns = group_by_frame.select_dtypes(include=['float64', 'int64']).columns
group_by_frame[numeric_columns].mean()
```
