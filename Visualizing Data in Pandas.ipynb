```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv(r"C:\AfterClass Learning\Python-Pandas\Data\Ice Cream Ratings.csv")
df = df.set_index('Date')
df
```

## Change the Style the library
- Check the available styles

```python
print(plt.style.available)
```

```python
plt.style.use('classic')
```

## Line plot

- Line plot is the default -> df.plot()
- If want to break the plots into subplots -> df.plot(kind = 'line', subplots =  True)
- Add titles as well -> df.plot(kind = 'line', subplots =  True,title = 'ICE CREAM RATING')
- lables can also be changed -> df.plot(kind = 'line', subplots =  True, title = 'ICE CREAM RATING', xlabel = 'Daily Rating', ylabel = 'Scores')

```python
df.plot()
```

```python
df.plot(kind = 'line', subplots =  True, title = 'ICE CREAM RATING', xlabel = 'Daily Rating', ylabel = 'Scores')
```

## Bar charts 

- Bar charts -> df.plot(kind = 'bar')
- for making stacked bar charts -> df.plot(kind = 'bar', stacked = 'True')
- if want a single value add -> df['Flavour Rating'].plot(kind = 'bar', stacked = 'True')

```python
df['Flavor Rating'].plot(kind = 'bar', stacked = 'True')
```

## Horizontal Bar charts  

```python
df.plot.barh()
```

## Scatter Plot
- Indorder to this plot to work x and y needs to be specified
- s is for the size of dots and c is for color

```python
df.plot.scatter(x = 'Texture Rating', y = 'Overall Rating', s = 300, c = 'red')
```

## Histogram
- Good for showing distributions
- Bins - num of bins histogram will use

```python
df.plot.hist(bins = 20)
```

## Boxplot
- Looking at the markers in the data

```python

df.boxplot()
```

## Area Plot

```python

df.plot.area(figsize = (10,5))
```

## Pie Chart
- We need to specify which coloumn currently working with
- figsize is changing the size

```python
df.plot.pie(y = 'Flavor Rating', figsize = (20,10))
```
